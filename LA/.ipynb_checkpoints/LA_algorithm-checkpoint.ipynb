### 박스 수 filtering까지 처리

```python
def filter_box(df, standard):
    filtered_b2c = df[(df[standard] >= 53.41)]
    filtered_b2b = filtered_b2c[(filtered_b2c[standard] >= 55.80)]
    return filtered_b2b
```

```python
df = raw_df.copy()
filtered_b2c_l = (raw_df['수량의 부피'] >= 53.41) & (raw_df['수량의 부피'] < 55.80)
filtered_b2b_l = (raw_df['수량의 부피'] >= 55.80)
filtered_combine_l = []
for i in zip(filtered_b2c_l, filtered_b2b_l):
    if i[0]:
        filtered_combine_l.append('*')
    elif i[1]:
        filtered_combine_l.append('**')
    else:
        filtered_combine_l.append('')
df['filtered'] = filtered_combine_l
df
```

```python
df['filtered'].value_counts()
```

```python
def filter_box(df, standard, b2c_standard, b2b_standard):
    df = raw_df.copy()
    filtered_b2c_l = (raw_df[standard] >= b2c_standard) & (raw_df[standard] < b2b_standard)
    filtered_b2b_l = (raw_df[standard] >= b2b_standard)
    filtered_combine_l = []
    for i in zip(filtered_b2c_l, filtered_b2b_l):
        if i[0]:
            filtered_combine_l.append('*')
        elif i[1]:
            filtered_combine_l.append('**')
        else:
            filtered_combine_l.append('')
    df['box filtered'] = filtered_combine_l
    return df
```

```python
raw_df_copy = raw_df.copy()
boxfiltered_df = filter_box(raw_df_copy, '수량의 부피', 53.41, 55.80)
boxfiltered_df 
```

```python
boxfiltered_df['box filtered'].value_counts()
```
